```python
#!/usr/bin/env python
import pyson
import pyson.runtime
import pyson.stdlib
import os
import random

# auctioneer parameters
'''english(0)
dutch(0)
winner(0)
auction(0)
rounds(0)
tttd(12)
tttdThreshold(10)
reservePrice(140)
priceIncrement(0.03)
dutchInitCoef(0.5)
winningBid(140)'''

# bidder parameters
'''auction(0)
reservePrice(150)
priceIncrement(0.03)'''

def runMassiveAuction(num_agents = 10):
    env = pyson.runtime.Environment()
    actions = pyson.Actions(pyson.stdlib.actions)

    priceIncrement = [a/100 for a in list(range(10, 30, 1))]
    b_reservePrices = [random.randrange(1000) for i in range(num_agents)]
    b_priceIncrements = [random.choice(priceIncrement) for i in range(num_agents)]
    


    with open("/home/maixent/Documents/MAS/Auctions/newtake/bidder_v5.asl") as source:
        agents = env.build_agents(source,num_agents,actions)
        
        for i in range(len(agents)):
            l=pyson.Literal("auction",(0,))
            pyson.runtime.add_belief(l,agents[i],pyson.runtime.Intention())
            l=pyson.Literal("reservePrice",(b_reservePrices[i],))
            pyson.runtime.add_belief(l,agents[i],pyson.runtime.Intention())
            l=pyson.Literal("priceIncrement",(b_priceIncrements[i],))
            pyson.runtime.add_belief(l,agents[i],pyson.runtime.Intention())
            
            print("agentName:",agents[i].name,"|","reservePrice:",b_reservePrices[i],"|","priceIncrement:",b_priceIncrements[i])
    
    a_tttd = random.randrange(200)
    a_tttdThreshold = random.randrange(200)
    a_reservePrice = random.randrange(1000)
    a_priceIncrement = random.choice(priceIncrement)
    a_dutchInitCoef = 3 * a_priceIncrement
    a_winningBid = a_reservePrice


    with open("/home/maixent/Documents/MAS/Auctions/newtake/auctioneer_vf.asl") as source:
        auctioneer=env.build_agent(source,actions)
        l=pyson.Literal("english",(0,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        l=pyson.Literal("dutch",(0,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        l=pyson.Literal("winner",(0,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        l=pyson.Literal("auction",(0,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        l=pyson.Literal("rounds",(0,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        l=pyson.Literal("tttd",(a_tttd,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        l=pyson.Literal("tttdThreshold",(a_tttdThreshold,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        l=pyson.Literal("reservePrice",(a_reservePrice,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        l=pyson.Literal("priceIncrement",(a_priceIncrement,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        l=pyson.Literal("dutchInitCoef",(a_dutchInitCoef,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        l=pyson.Literal("winningBid",(a_winningBid,))
        pyson.runtime.add_belief(l,auctioneer,pyson.runtime.Intention())
        agents.append(auctioneer)
        
    print("tttd",a_tttd, "a_tttdThreshold",a_tttdThreshold, "a_reservePrice",a_reservePrice, "a_priceIncrement",a_priceIncrement, "a_dutchInitCoef",a_dutchInitCoef,"a_winningBid",a_winningBid)
    env.run()
```

```python
runMassiveAuction(2)
```
